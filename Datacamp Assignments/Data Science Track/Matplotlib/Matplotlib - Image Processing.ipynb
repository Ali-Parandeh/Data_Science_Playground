# Image Processing with Matplotlib

```python
# Importing the required packages
from dateutil import parser
from time import time
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import math

# Generate the plots and charts within this window instead of opening a new window for each chart/plot
%matplotlib inline
```

```python
img = plt.imread("flower.jpg")
print(img.shape)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
collapsed = img.mean(axis = 2)
print(collapsed.shape)
plt.imshow(collapsed)
plt.show()
```

```python
collapsed = img.mean(axis = 2)
print(collapsed.shape)
plt.imshow(collapsed, cmap = "gray")
plt.show()
```

```python
uneven = collapsed[::4, ::2] # nonuniform sample -  ::4 means sample every 4th element. 3818/4 = 955. 5425/2 = 2713
plt.set_cmap("gray")
print(uneven.shape)
plt.imshow(uneven);
```

```python
uneven = collapsed[::4, ::2] # nonuniform sample 
# The reason image above was skewed was due to incorrect assumptions about the aspect ratio, that is each pixel size is 1x1.
# Here we need to specify that now the pixel sizes are 2x1 due to out non-uniform sampling of the photo.
print(uneven.shape)
plt.imshow(uneven, aspect = 2.0); # Pay attention to the y-axis values. I have literally reduced the resolution.
```

```python
uneven = collapsed[::4, ::2] # nonuniform sample
print(uneven.shape)
# Defines x and y axis limits and resizes pixels accordingly. Here the aspect ratio is defined using these limits for pixels
plt.imshow(uneven, extent=(0, 800, 0, 600)); 
```
